```python
from scipy.optimize import curve_fit

def func_QE_term(T_ex, a, b, c, d):
    if isinstance(T_ex, u.Quantity):
        T_ex = T_ex.cgs.value
    return a + b * T_ex + c * np.exp(-T_ex) + d * np.exp(-T_ex**0.5) # * (c / T_ex**d)

def fit_func_QE_CO(x, y):
    popt, pcov = curve_fit(func_QE_term, x, y, p0=[1.0, 1.0, 1.0, 0.2])
    yfit = func_QE_term(x, *popt)
    return popt, yfit

popt, yfit = fit_func_QE_CO(x, y)

fig, ax = plt.subplots(1, 1)
ax.plot(x, y)
ax.set_xscale('log')
ax.xaxis.set_major_formatter(mpl.ticker.FormatStrFormatter('%g'))
ax.plot(x, yfit, ls='dashed')
print('popt:', repr(popt))
```
